```python
# Mock Tesla stock data
tesla_stock_data = {
    "Date": ["2023-03-31", "2023-02-28", "2023-01-31", "2022-12-31", "2022-11-30"],
    "Open": [200, 195, 190, 180, 170],
    "High": [210, 200, 195, 185, 175],
    "Low": [195, 185, 180, 170, 160],
    "Close": [205, 190, 185, 175, 165],
    "Adj Close": [205, 190, 185, 175, 165],
    "Volume": [30000000, 25000000, 20000000, 15000000, 10000000]
}
tesla_data = pd.DataFrame(tesla_stock_data)
tesla_data['Date'] = pd.to_datetime(tesla_data['Date'])
tesla_data.set_index('Date', inplace=True)
print("Tesla data created successfully.")
tesla_data.head()
```

```python
# Mock Tesla revenue data
tesla_revenue_data = {
    "Date": ["2023-03-31", "2022-12-31", "2022-09-30", "2022-06-30", "2022-03-31", 
             "2021-12-31", "2021-09-30", "2021-06-30", "2021-03-31", "2020-12-31"],
    "Revenue": [18400, 24300, 21400, 16934, 10389, 
                17719, 13957, 11958, 10789, 10744]
}
tesla_revenue = pd.DataFrame(tesla_revenue_data)
tesla_revenue['Date'] = pd.to_datetime(tesla_revenue['Date'])
tesla_revenue.set_index('Date', inplace=True)
print("Tesla revenue data created successfully.")
tesla_revenue.tail()
```

```python
# Plot stock data
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data['Close'], label='Close Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    plt.show()

# Plot Tesla stock graph
make_graph(tesla_data, "Tesla Stock Price (2023)")

# Plot GameStop stock graph
make_graph(gme_data, "GameStop Stock Price (2023)")
```
